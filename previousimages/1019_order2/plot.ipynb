```python
import matplotlib.pyplot as plt
from matplotlib.pyplot import imread

# Read your PNG files
image1 = imread('my_target.png')
image2 = imread('naive_without_ed.png')
image3 = imread('naive_with_ed.png')
image4 = imread('ours_without_ed.png')
image5 = imread('ours_with_ed.png')

# Create a figure and a 1x2 grid of subplots
fig, axes = plt.subplots(1, 2, figsize=(15, 5))

# Plot the first image on the left (larger)
axes[0].imshow(image1)
axes[0].set_title('Pure Watermark')

# Create a 2x2 grid of subplots on the right
gs = fig.add_gridspec(2, 2, wspace=0.2, hspace=0.2, width_ratios=[1, 1])

# Plot the other four images on the right
ax2 = fig.add_subplot(gs[0, 0])
ax2.imshow(image2)
ax2.set_title('Naive, without ed')

ax3 = fig.add_subplot(gs[0, 1])
ax3.imshow(image3)
ax3.set_title('Naive, with ed')

ax4 = fig.add_subplot(gs[1, 0])
ax4.imshow(image4)
ax4.set_title('Ours, without ed')

ax5 = fig.add_subplot(gs[1, 1])
ax5.imshow(image5)
ax5.set_title('Ours, with ed')

# Adjust the spacing between subplots
plt.subplots_adjust(wspace=0.3)

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
from matplotlib.pyplot import imread

# Read your PNG files
image2 = imread('ring_naive_without_ed_cut_order2.png')
image3 = imread('ring_naive_with_ed_cut_order2.png')
image4 = imread('ring_ours_without_ed_cut_order2.png')
image5 = imread('ring_ours_with_ed_cut_order2.png')

# Create a figure and a 2x2 grid of subplots
fig, axes = plt.subplots(2, 2, figsize=(10, 10))

# Plot the other four images on the right
axes[0, 0].imshow(image2)
axes[0, 0].set_title('Naive, without ed')

axes[0, 1].imshow(image3)
axes[0, 1].set_title('Naive, with ed')

axes[1, 0].imshow(image4)
axes[1, 0].set_title('Ours, without ed')

axes[1, 1].imshow(image5)
axes[1, 1].set_title('Ours, with ed')

# Adjust the spacing between subplots
plt.subplots_adjust(wspace=0.3, hspace=0.1)

# Show the plot
plt.show()
```
